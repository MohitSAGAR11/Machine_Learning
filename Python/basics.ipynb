Python Control Flow


```
# if/else/elif statement

age = 18
if age > 18:
    print("valid")
elif age == 18:
    print("Perfect")
else:
    print("invalid")
```

```
## Nested conditional statement
num = int(input("Enter a number"))

if num>0:
    print("Positive")
    if num%2==0:
        print("Even")
    else:
        print("Odd")
else:
    print("Negative or Zero")
```

```
## Loops in python

## for loop

# for i in range(10 , 1 , -1): # (start , end , stepSize)
#     print(i)


str = "KrishNaik"

for i in str:
    print(i)
```

```
## while loop

count = 0
while count > 10:
    if count % 2 == 0:
        break
    print(count)
    count += 1
```

```
# Break : Break the loop when conditions met
# Continue : Skips the current iteration
# Pass : null operation , it does nothing

for i in range(5):
    if (i % 2) == 0:
        pass
    else:
        print(i)
```
